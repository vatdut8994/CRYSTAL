```python
import requests
import json

user_query = input("Enter your search query: ")
api_url = "http://localhost:3001/api/search"
headers = {'Content-Type': 'application/json'}
data = json.dumps({
  "chatModel": {
    "provider": "groq",
    "model": "mixtral-8x7b-32768"
  },
  "embeddingModel": {
    "provider": "local",
    "model": "xenova-bge-small-en-v1.5"
  },
  "optimizationMode": "balanced",
  "focusMode": "webSearch",
  "query": user_query,
  "history": []
})

response = requests.post(api_url, headers=headers, data=data)
result = response.json()

print("Result:")
print(json.dumps(result, indent=4))
```

```python
from moviepy.editor import VideoFileClip

def convert_mov_to_wav(mov_file):
    """
    Convert a .mov file to a .wav file.
    
    Parameters:
    mov_file (str): Path to the .mov file.
    
    Returns:
    str: Path to the converted .wav file.
    """
    video = VideoFileClip(mov_file)
    audio = video.audio
    wav_file = mov_file.replace('.mov', '.wav')
    audio.write_audiofile(wav_file)
    return wav_file

# Example usage:
mov_file = '/Users/vatsaldutt/Downloads/ScreenRecording_01-14-2025 08-49-25_1.mov'
wav_file = convert_mov_to_wav(mov_file)
print(f"Converted {mov_file} to {wav_file}")
```
